## Rappels sur la notion d'entropie de Shannon dans le cas discret:

L'entropie d'un source aléatoire discrète X à valeurs dans $\mathcal{X} = {\{x_i\}}_{i=1..N}$, telle que définie par Shannon, se calcule par :
$$
H(X) = -\sum_{i=1}^{N}p(X=x_i)\times log_b p(X=x_i) = -\sum_{i=1}^{N}p_i\times log_b\ p_i 
$$
Où l'on note $p_i = p(X=x_i) \forall i \in [1,N]$, et $log_b(x)=\frac{ln(x)}{ln(b)}$ est le logarithme en base b (souvent base $b=2$ pour des variables binaires en informatique).

L'entropie d'une source donne une mesure de l'information moyenne apportée par chaque nouvelle observation de cette source pour un observateur. Si la variable observée est constante, par exemple $p_i=1$ et $p_j=0\ \forall j\neq i$, l'entropie est nulle puisque on apprend jamais rien de nouveau avec cette source.  
  
**Théorème d'entropie maximale :** L'entropie est maximale pour une source avec une distribution de probabilité uniforme : $\forall i\ :\ p_i=\frac{1}{N}$.  
Voir preuve et question bonus à la fin.  
  
Ce théorème montre qu'une source de distribution uniforme donne une maximum d'information à un observateur: ceci signifie qu'un observateur de X est à chaque moment surpris au maximum si X suit une distribution uniforme. En effet si il existait une probabilité $p_i$ supérieure aux autres d'observer un certain $X=x_i$, alors l'observateur serait un peu moin surpris en observant $X=x_i$, et ce d'autant plus que $p_i$ serait élevée.

## Liens entre entropie de Shannon et codage de symboles

On peut donner une explication pratique à la forme de la fonction entropie $H(X)=-\sum_{i=1}^{N}p_i\times log_b\ p_i$.  
  
Considérons un ensemble de catégories de symboles. Les catégories au nombre de $C$ sont notées $x^i, i\in[1,C]$, et contiennent chacune $N_i$ symboles différents $s^i_j, i\in[1,C], j\in[1,N_i]$. Notons alors $N = \sum_i^{C} N_i = |\{s^i_j\}|$ le nombre total de symboles possibles, toute catégorie confondues.  
  
Supposons que l'on observe alors des symboles émis par une source aléatoire $S$, avec une probabilité uniforme d'observer chaque symbole :$\forall(i,j)\ p(S=s^i_j)=\frac{1}{N}$. On a alors $p(X=x^i)=\sum_{j=1}^{N_i} p(S=s^i_j)=\frac{N_i}{N}$ la probabilité que le symbole S soit de la catégorie $i$. On peut imaginer comme source une lettre de l'alphabet observée aléatoirement et comme catégories les consomnes ($N_i=20$) et les voyelles ($N_i=6$).  
  
Pour transmettre l'information du symbole en code binaire, il faut un nombre de bits suffisant pour coder le symbole: si on encode le symbole par un numéro, le nombre de bits pour coder ce numéro parmis les $N$ possibilités est exactement égal à $log_2(N)$. Par exemple si il y a 16 symboles possibles, il faudra 4 bits pour encoder les numéros des symboles.
On peut aussi découper le code en deux partie : coder la catégorie puis le symbole dans la catégorie i. Connaissant une catégorie de taille $N_i$, on peut coder le numéro d'un symbole au sein de cette catégorie avec $log_2(N_i)$ bits. Il reste alors à déterminer le nombre de bits nécéssaires pour coder la catégorie.    
  
Notre inconnue est donc ici la taille du code permettant d'encoder la catégorie. Notons X la variable aléatoire du numéro de cette catégorie, et notons $H(X)$ la taille du code nécéssaire pour encoder sa valeur. On a alors :
$$
log_2(N) = H(X) + moyenne(log_2(N_i)) = H(x) + \sum_{i=1}^{C}p(X\in x^i)log_2(N_i)
$$
Ce qui donne :
$$
-H(x) = \sum_{i=1}^{C}p(X\in x^i)log_2(N_i) - log_2(N)\; \iff\; -H(x) = \sum_{i=1}^{C}\frac{N_i}{N}log_2(\frac{N_i}{N})\; \iff\; H(x) = -\sum_{i=1}^{C}p_i \times log_2(p_i)
$$
On avait supposé que tous les symboles avaient une probabilité égale d'apparaitre, mais comme on les a groupé par catégorie et que l'on a regardé la catégorie, on a du insérer leur probabilité d'appartenance à ces catégories. Finalement cet exemple permet de comprendre comment fonctionne l'entropie d'une source pour laquelle on a des probabilités différentes pour les différentes valeurs.

## Encodage des pixels avec le code de Huffman
On suppose ici connaître la dimension des images (donc leur nombre de pixels et leur nombre de lignes), et on souhaite encoder ces images en séquences binaires les plus compactes possibles.
On va pour cela utiliser le codage de Huffman.

### Question 2
Compléter le code de la classe ```noeud_huffman``` qui crée un noeud d'un arbre de Huffman.

````python
class Noeud_huffman:
    def __init__(self, valeur=None):
        # code binaire (pour tous les noeuds sauf la racine)
        self.bit = None
        # valeur (pour les noeuds feuilles uniquement)
        self.valeur = valeur
        # noeud parent (pour tous les noeuds sauf la racine)
        self.parent = None
        # noeuds enfants (pour tous les noeuds sauf les feuilles)
        # devra être une liste de deux noeuds
        self.enfants = None
        
    def remonte_arbre(self, code=[]):
        # cette fonction récurrent permet de remonter l'arbre d'une feuille à la racine et de retourner le code        
        if not self.parent:
            return code
        else:
            code = [self.bit] + code # ajoute le bit du noeud en tête du code actuel
            return self.parent.remonte_arbre(code)
        
    def descend_arbre(self, code, profondeur=0):
        # cette fonction récurrente permet de descendre l'arbre de la racine à une feuille et de retourner
        # la valeur de la feuille ainsi que la profondeur de la feuille dans l'arbre (longueur du code)
        if not self.enfants:
            # si le noeud n'a pas d'enfants, alors c'est une feuille avec une valeur
            return self.valeur, profondeur
        else:
            # sinon on regarde le premier bit du code et on descend dans l'enfant correspondant
            return self.enfants[code[0]*1].descend_arbre(code[1:] if len(code>1) else [], profondeur+1)
        
    def descend_arbre_iteratif(self, code):
        # cette fonction iterative permet de descendre l'arbre de la racine à une feuille et de retourner
        # la valeur de la feuille ainsi que la profondeur de la feuille dans l'arbre (longueur du code)
        noeud = self
        profondeur=0
        while noeud.enfants is not None:
            noeud = noeud.enfants[code[profondeur]*1]
            profondeur+=1
        return noeud.valeur, profondeur
        
````

### Question 3
Compléter le code de la fonction ```__init__``` de la classe ```Arbre_huffman``` qui crée l'arbre d'une image.

````python
class Arbre_huffman:
    def __init__(self, image):        
        # calcule des probas pour chaque valeur de gris
        #flatim = image.flatten()
        #valeurs_possibles = list({ pixel for pixel in flatim }) # liste de l'ensemble de toutes les valeurs possibles de pixels
        #occurences_pixel = np.array([ sum(flatim==pixel) for pixel in valeurs_possibles])        
        #p = occurences_pixel/len(flatim)
        # version rapide
        p = np.histogram(image,256)/np.double(image.size)
        valeurs_possibles = array([ i for i in range(p.size) if p[i]>0 ])
        
        # on crée un tableau qui associe chaque valeur à un noeud feuille vierge, puis qui sera mis à jour
        # avec les noeuds en cours de construction
        tableau_noeuds = [Noeud_huffman(valeurs_possibles[k]) for k in range(len(valeurs_possibles)) ]

        # idem pour les probas des noeuds en cours de construction
        p_noeuds = p.tolist()

        # on garde en mémoire un dictionnaire qui associe à chaque valeur possible un noeud feuille:
        dico_feuilles = { valeurs_possibles[k]:tableau_noeuds[k] for k in range(len(valeurs_possibles))}

        # on va iterativement créer un noeud parent pour les deux noeuds de plus petite proba
        # puis sommer ces probas pour obtenir la proba du noeud parent et mettre les deux tableaux à jour
        for i in range(len(p)-1):

            # tri des probas par ordre croissant, on récupère les indices du tri de p_noeuds
            # indices_p_tri[0] est donc l'indice du minimum de p_noeuds
            indices_p_tri = sorted(range(len(p_noeuds)), key=lambda k: p_noeuds[k])        
            # on assigne les bits à 0 et 1 aux deux noeuds de plus petites proba
            tableau_noeuds[indices_p_tri[0]].bit = 0
            tableau_noeuds[indices_p_tri[1]].bit = 1
            # on crée un noeud parent, et on assigne ce parent aux enfant, et vice-versa
            noeud_parent = Noeud_huffman()
            tableau_noeuds[indices_p_tri[0]].parent = noeud_parent
            tableau_noeuds[indices_p_tri[1]].parent = noeud_parent
            noeud_parent.enfants = [ tableau_noeuds[indices_p_tri[0]], tableau_noeuds[indices_p_tri[1]] ]
            # on somme ensuite les probas des deux noeuds
            somme_probas = p_noeuds[indices_p_tri[0]] + p_noeuds[indices_p_tri[1]]
            # on supprime les deux noeuds enfant du tableau ainsi que leur probas :
            del tableau_noeuds[max(indices_p_tri[0:2])]
            # on supprime dans l'ordre le noeud d'indice le plus grand plus l'autre à cause du décalage des
            # indices après suppression d'un indice
            del tableau_noeuds[min(indices_p_tri[0:2])] 
            # idem pour le tableau de probas des noeuds :
            del p_noeuds[max(indices_p_tri[0:2])]
            del p_noeuds[min(indices_p_tri[0:2])]
            # on ajoute ensuite à ces deux tableaux le nouveau noeuds et sa proba somme :
            tableau_noeuds.append(noeud_parent)
            p_noeuds.append(somme_probas)
            # ces deux tableaux sont maintenant de taille len(p)-i

        # à la sortie de la boucle il ne reste qu'un seul noeud dans le tableau, la racine de notre arbre
            # qui est de proba 1 et ne contient pas de valeur 0 ou 1 (seul noeud sans valeur)
        # on sauvegarde le dico des feuilles qui peuvent maintenant être utilisées pour 
            # pour coder en remontant jusqu à la racine (le seul noeud sans parent)
        self.feuilles = dico_feuilles
        # on sauvegarde aussi le noeud racine qui permet de décoder en descendant jusqu'à
            # une feuille (seuls noeuds sans enfant)
        self.racine = tableau_noeuds[0]
        
    def encode_valeur(self, valeur):
        code = self.feuilles[valeur].remonte_arbre()
        return np.array( code, dtype='bool' )

    def encode_image(self, image):
        flatim = image.flatten()
        code = []
        for k in range(len(flatim)) :
            code += [self.encode_valeur(flatim[k])]
        return np.concatenate(code)
        
    def decode_valeur(self, code):
        # on va descendre depuis la racine jusqu'à une feuille et renvoyer sa valeur
        # on renvoie aussi la longueur du code de cette valeur
        return self.racine.descend_arbre(code)
    
    def decode_valeur_iteratif(self, code):
        # on va descendre depuis la racine jusqu'à une feuille et renvoyer sa valeur
        # on renvoie aussi la longueur du code de cette valeur
        return self.racine.descend_arbre_iteratif(code)

    def decode_image(self, code, shape):
        image = np.zeros([shape[0]*shape[1],1])
        index_code = 0
        for i in range(shape[0]*shape[1]):            
            image[i], profondeur = self.decode_valeur(code[index_code:])
            index_code += profondeur
            if i%shape[0]==0:
                print('decodage ligne ' + str(int(i/shape[0])) + ' sur ' + str(shape[0]) )
        return np.reshape(image, shape)
    
    def decode_image_iteratif(self, code, shape):
        image = np.zeros([shape[0]*shape[1],1])
        index_code = 0
        for i in range(shape[0]*shape[1]):            
            image[i], profondeur = self.decode_valeur_iteratif(code[index_code:])
            index_code += profondeur
        return np.reshape(image, shape)
            
````

### Question 7 (Bonus)
Créer une fonction ```descend_arbre_recurrente``` qui descend l'arbre de façon récurrente et une autre fonction ```descend_arbre_iteratif```qui descend l'arbre de façon iterative. Utilisez les pour décoder une image et comparez leurs vitesses. Laquelle est la plus rapide, pourquoi ? 

#### Réponse :
La fonction itérative est bien plus rapide car elle évite de produire des piles d'appel de fonctions coûteuse en temps et mémoire, surtout en Python


**Preuve du théorème d'entropie maximale :** Considérons la distribution $\forall i\in[1..N]:\ p(X=x_i)=p_i$, avec $\sum_{i=1}^{N}p_i\ =\ 1$.  
La variable X prend des valeurs discrètes, mais les probabilités $p_i$ peuvent évoluer de manière continue d'une distribution a une autre. On cherche alors la distribtion $p=(p_1,...,p_N)$ qui maximise l'entropie $H_{X\sim p}(X)=-\sum_{i=1}^{N}p_i\times log_b\ p_i$ sous la contrainte d'avoir $\sum_{i=1}^{N}p_i\ =\ 1$, ce qui s'écrit :
$$
\max_p\ H_{X\sim p}(X)=-\sum_{i=1}^{N}p_i\times log_b\ p_i\ \ \ s.c.\ \ \sum_{i=1}^{N}p_i\ =\ 1
$$
On cherche à trouver maximum sous une contrainte, on doit donc chercher le point selle du Lagrangien de l'équation :
$$
\min_{\lambda}\max_{p}\ L_(p,\lambda)\ =\ -\sum_{i=1}^{N}p_i\times log_b\ p_i\ \ - \ \lambda(\sum_{i=1}^{N}p_i\ -\ 1)
$$
On cherche alors $p$ et $\lambda$ qui annulent le gradient de ce Lagrangien. Le gradient par rapport à $\lambda$ est:
$$
\nabla_\lambda L_(p,\lambda) = \sum_{i=1}^{N}p_i\ -\ 1
$$
qui s'annule sans étonnement pour $p$ tel que $\sum_{i=1}^{N}p_i\ =\ 1$.  
  
Le gradient par rapport à p est :
$$
\forall i\ :\ \nabla_{p_i}L_(p,\lambda)\ =\ -log_b(p_i) - \frac{p_i}{p_i.ln(b)} - \lambda \ =\ -log_b(p_i) - \frac{1}{ln(b)} - \lambda
$$
et s'annule pour les $p_i$ tels que :
$$
\forall i\ : \lambda = -log_b(p_i) - \frac{1}{ln(b)} \;\iff\;\forall i\ : p_i = b^{-\frac{1}{ln(b)}-\lambda}
$$
On injecte cela dans l'équation précédente :
$$
\sum_{i=1}^{N}p_i\ =\ 1\;\iff\; \sum_{i=1}^{N}b^{-\frac{1}{ln(b)}-\lambda} = 1 \;\iff\; N.b^{-\frac{1}{ln(b)}-\lambda} = 1 \;\iff\; \lambda = -\frac{1}{ln(b)}-log_b(\frac{1}{N})
$$
Et finalement comme on avait $\lambda = -log_b(p_i) - \frac{1}{ln(b)}$ :
$$
