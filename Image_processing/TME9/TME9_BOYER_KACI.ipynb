Recopier ci-dessous le code des fonctions ``split()``, ``predsplit1()`` et ``splitprof()``
écrites au TME 8:

````python
def split(I,reg,pred,*args):
    """ Array*Bloc*(Array*Bloc*...->bool)*... -> 4-aire
        Découpage de I en arbre 4-aire sur un critère d'écart-type minimal
    """
     #n vertical; m horizontal
    x,y,n,m=reg
    if(pred(I,reg,* args) and n>2 and m>2):
        n2=n//2
        m2=m//2
        NO=split(I,(x,y,n2,m2 ),pred,*args)
        NE=split(I, (x, y+m2, n2, m2), pred, *args)
        SO=split(I, (x+n2, y, n2, m2), pred, *args)
        SE=split(I, (x+n2, y+m2, n2, m2), pred, *args)
        return [NO,NE,SO,SE]
    else:
        return [reg]

def predsplit1(I,reg,*args):
    """ Array*Bloc*... -> bool
    """
    x,y,n,m=reg
    return I[x:x+n,y:y+m].std()>args[0]
 
def splitprof(L):
    """ 4-aire -> list[Bloc]"""
    #la liste de retour
    if(len(L)==1):
        return L
    else:
        return splitprof(L[0])+splitprof(L[1])+splitprof(L[2])+splitprof(L[3])

````

### Question 1
Modifier la fonction ``splitprof(L)`` pour en faire une fonction ``splitprofstat(L,I)`` qui rend une liste de feuilles de l'arbre 4-aire ``L`` et telle que chaque élément de cette liste, de type ``Bloc``, est enrichi de trois scalaires, le premier étant la taille, le second la moyenne et la troisième la variance des valeurs de pixels du bloc dans l'image ``I``. Cette fonction rend alors une liste dont les éléments ont pour type formel:
```python
type Blocstat = tuple[int**4,int,float**2]
```
Les quatre premières valeurs sont celles du type ``Bloc``, la cinquième est la taille du bloc (en nombre de pixels) et les deux dernières correspondent à la moyenne et la variance calculées sur la région.


````python
# type Blocstat = tuple[int**4,int,float**2]

def splitprofstats(L,I):
    """ 4-aire*Array -> list[Blocstat] """
    if(len(L)==1):
        x, y, n, m=L[0]
        return [(x, y, n, m, n*m, I[x:x+n,y:y+m].mean(),I[x:x+n,y:y+m].std()**2)]
    else:
        return splitprofstats(L[0], I)+splitprofstats(L[1], I)+splitprofstats(L[2], I)+splitprofstats(L[3], I)
````

### Question 2
Dans la suite du TME, on considère le type formel:
```python
type Region = list[Blocstat]
```
Une région, comme vu au TME, est donc une liste de blocs. Écrire le prédicat de fusion ``predmerge(b,R,*args)`` vu en TD. Cette fonction retourne ```True``` si le bloc ``b`` doit fusionner dans la région ``R``. Si telle est le cas, alors le premier élément de la liste ``R`` voit ses statistiques modifiées: elles décrivent celles de l'ensemble des blocs de la région ainsi fusionnée.

````python
def predmerge(b,R,*args):
    """ Blocstat*Region*... -> bool 
        Si fusion, R[0] est modifié
    """
    x1, y1, n1, m1, size1, moy1, var1=b
    x2, y2, n2, m2, size2, moy2, var2= R[0]
    size=size1+size2
    moy=(size1*moy1+size2*moy2)/size
    # formule calculée en TD
    var=(size1*var1+size2*var2+size1*moy1*moy1+size2*moy2*moy2)/size-moy*moy
    if(var<args[0]):
        R[0]=x2, y2, n2, m2, size, moy, var
        return True
    return False
````

### Question 3
Implémenter les fonctions ```estdans()```, ```sontvoisins()```, ```voisins()``` et finalement ```merge()``` vues en TD.

````python
def estdans(x,y,a):
    """ int*int*Blocstat -> bool 
        Teste si (x,y) est dans a 
    """
    return x>=a[0] and x<a[0]+a[2] and y>=a[1] and y<a[1]+a[3] 

def sontvoisins(a,b):
    """ Blocstat**2 -> bool
        Teste si a et b sont voisins
    """
    
    if (a[2]>b[2] and a[3]>b[3]):
        a,b=b,a
    # a plus petit que b
    #les coordonnées du milieu de la région
    x,y=a[0]+a[2]//2 ,a[1]+(a[3]//2)
    return estdans(x,y+a[3],b) or estdans(x,y-a[3],b) or estdans(x-a[2],y,b) or estdans(x+a[2],y,b)
    

def voisins(b,K):
    """ Blocstat*list[Blocstat]->list[Blocstat]
        retourne la liste des elements de K voisins de b
    """
    L=[]
    for e in K:
        if sontvoisins(b, e):
            L.append(e)
    return L
# type Region = list[Blocstat]

#Entrée: S: liste des feuilles du quadtree, retourné par profstats suite à un split, donc on n'a pas besoin d'ajouter les stats dans la fonction merge, cela a déja été fait dans profstats
#        I: image
#       pred: predicat qui définit le critère de fusion
def merge(S,I,pred,*args):
    """ 4-aire*Array*(Blocstat*Region*...->bool) -> list[Region]
        Fusionne les feuilles de S en une liste de régions
    """
    #liste des feuilles avec les nouvelles statistiques
    K=splitprofstats(S,I)
    L=[]
    while(len(K)>0):
        #on enlève le premier élément de la liste des feuilles
        b=K[0]
        K=K[1:]
        #Nouvelle région
        R=[b]
        N=voisins(b,K)
        while(len(N)>0):
            bi=N[0]
            N=N[1:]
            if(pred(bi, R,* args)):
                K.remove(bi)
                R.append(bi)
                for j in voisins( bi, K):
                    if not j in N:
                        N.append(j)
        L.append(R)
    return L

def general_split_merge(I,seuil_split, seuil_merge):
    n,m=shape(I)
    res=split(I,(0,0,n,m),predsplit1, seuil_split)
    #LL=splitprofstats(res, I)
    L=merge(res, I, predmerge, seuil_merge)
    print( "nb blocs="+str(shape(L)[0]))
````

### Question 5
Écrire une fonction ``regions(LR,shape)`` qui prend une liste de régions telle que rendu par la fonction ``merge()`` et rend une image des régions. Chaque région sera colorée avec le niveau de gris correspondant à la moyenne de la région. Le paramètre ``shape`` donne la taille de l'image à produire. Tester la fonction sur l'exemple précédent.

````python
def regions(LR,shape):
    """ list[Region] -> Array """
    buf = zeros(shape,dtype=uint8)
    for R in LR:
        x,y,n,m,s,mu,v=R[0]
        for b in R:
            x,y,n,m,s,t,v=b
            buf[x:x+n,y:y+m] = int(mu)
    return buf
````

````python
# fonction pour appliquer toute les fonctions précédentes d'un seul coup
# retourne l'image et le nombre de blocs  après le merge
def general_split_merge_affich(I,seuil_split, seuil_merge):
    n,m=shape(I)
    res=split(I,(0,0,n,m),predsplit1, seuil_split)
    #LL=splitprofstats(res, I)
    print("nb blocs au split "+str(shape(splitprofstats(res, I))[0]))
    L=merge(res, I, predmerge, seuil_merge)
    buff=regions(L, shape(I))
    #plt.axis('off')
    plt.title('seuil split='+str(seuil_split)+'seuil merge'+str(seuil_merge))
    print( "nb blocs au merge="+str(shape(L)[0]))
    imshow(buff, cmap='gray')
    return buff, shape(L)[0]
````

### Question 2 (bonus):
Utiliser un critère de découpage alternatif, par exemple celui proposé en cours: on fusionne qui la variance de la région fusionné est plus petite que celle de la région non fusionnée. Tester sur les images.

````python
def predmerge2(b,R,*args):
    """ Blocstat*Region*... -> bool 
        Si fusion, R[0] est modifié
    """
    x1, y1, n1, m1, size1, moy1, var1=b
    x2, y2, n2, m2, size2, moy2, var2= R[0]
    size=size1+size2
    moy=(size1*moy1+size2*moy2)/size
    # formule calculée en TD
    var=(size1*var1+size2*var2+size1*moy1*moy1+size2*moy2*moy2)/size-moy*moy
    if(var<max(var1,var2)):
        R[0]=x2, y2, n2, m2, size, moy, var
        return True
    return False
````
